```python
import pandas as pd
from datetime import datetime
```

```python
datelist = pd.date_range(datetime.today(), periods=31).tolist()
```

```python
numbers = [43,3,31,1,39,18,15,49,6,14,46,45,15,13,5,4,23,25,43,4,18,3,4,14,3,42,53,2,1,35,14]
```

```python
len(datelist)
```

```python
len(numbers)
```

```python
datelist
```

```python
df= pd.DataFrame({'Dates':datelist,'Price':numbers})
```

```python
df.head(10)
```

```python
df.plot.line(x='Dates',y='Price')
```

```python
df['Moving Average']=df['Price'].rolling(window=5).mean()
```

```python
df['Moving Average'].plot.line(x='Dates',y='Price')
```

```python
# or like this...

df.plot.line(x='Dates',y=['Price','Moving Average'])
```

```python
df['Rolling Average Center']=df['Price'].rolling(window=5, center=True).mean()
```

```python
df['Rolling Average Center']

#now it uses values 0-4 to calculate 2
```
