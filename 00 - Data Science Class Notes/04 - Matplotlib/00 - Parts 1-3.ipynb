## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))#,dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)


ax.set_title('TITLE')
ax.set_ylabel('Y')
ax.set_xlabel('X')
#figsize=(width x height inches)
#dpi dots/pixels per inch
```

#### Now with subplots

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize = (8,2))


axes[0].plot(x,y)
axes[1].plot(y,x)
#figsize=(width x height inches)
#dpi dots/pixels per inch
plt.tight_layout
```

### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)


ax.plot(x,x**3)

#just like this, we have unlabled lines!
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='BUTTS')

ax.plot(x,x**3,label='TOOTS')

ax.legend(loc=0)
```

## Customizing Options

### colors

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'orange',linewidth=7,
        alpha=.5,linestyle='-',marker='o',markersize=10,
       markerfacecolor='black',markeredgewidth=10,markeredgecolor='pink')

# can be ==== color = RGB HEX code
#linewidth 1 is standard. 2 is doubly think etc.
    #abreviated lw
#alpha how transparent 1= perfect solid. 0= invis
#linestyle = dashed, dotted etc.
    #abreviated ls
    
#markers.
    #exact data points
```

### Axes apearance and plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'purple',lw=2,ls='--')

#what if just want 0-1 on x?
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
