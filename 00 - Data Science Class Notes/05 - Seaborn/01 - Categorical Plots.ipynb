# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Barplot

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Countplot
same as barplot but estimator always counts number of occurances. just x value

```python
sns.countplot(x='sex',data=tips)

#there is more males in data set than females
```

### Box (&whisker) plots 

```python
#shows dif in quantitative vars to highlight dif
#lil' diamonds are outliers


#shows quarters (the T's and the colored boxes.)
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')

#smokers spend more on weekends??
```

### Violin plot

shows dist acros category
x is cat
y is num

```python
sns.violinplot(x='day',y='total_bill',data=tips)

#great to show distribution. more info but harder to read
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### Strip plot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)

#jitter adds random noise to make more clear
```

### Swarm plot
points are adjusted so they don't overlap.
combo of a strip plot and violin plot
Drawback. Don't scale well to large numbers

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#lets plot on top of eachother
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

### Factor plot

takes in x and y. 
Kind arg = describes plot

```python
#you can use this for all. then just change kind to 'bar' or 'violin' etc.
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
